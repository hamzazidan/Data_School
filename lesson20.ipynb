# When should I use the "inplace" parameter in pandas?


```python
import pandas as pd
```

```python
ufo = pd.read_csv("http://bit.ly/uforeports", 
                  index_col=['City', 'Colors Reported', 
                             'Shape Reported', 'State',
                             'Time'])
```

```python
ufo.head()
```

```python
ufo.reset_index()
```

```python
ufo.drop('City', axis=1)
```

```python
ufo.head(5)
```

```python
ufo.drop(columns='City')
```

```python
ufo.head()
```

```python
ufo.drop(columns='City', inplace=True)
```

```python
ufo.head()
```

```python
ufo.dropna(how='any').shape
```

```python
ufo.set_index("Time", inplace=True)
ufo
```

```python
ufo.fillna(method='bfill')
```
