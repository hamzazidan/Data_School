## How do I apply multiple filter criteria to a pandas DataFrame

```python
import pandas as pd
```

```python
movies = pd.read_csv("http://bit.ly/imdbratings")
```

```python
movies.head()
```

```python
movies[(movies['duration'] >= 200) & (movies['star_rating'] > 9)]
```

```python
movies[(movies['duration'] >= 200) | (movies['genre'].isin(['Drama', 'Action']))]
```

```python
movies['genre'].isin(['Drama', 'Action'])
```

```python
movies[movies['genre'].isin(['Drama', 'Action', 'Adventure'])]
```

```python
import numpy as np
movies.select_dtypes(include=[np.number]).dtypes
```

```python
numeric_values = movies.select_dtypes(include=[np.number])
```
